다음은 5개 지점에서 판매된 과일에 대한 월별 판매량 데이터입니다. 이 데이터는 각 지점의 판매량을 월별로 나타내며, 각 지점에서 판매된 사과, 바나나, 오렌지의 판매량이 기록되어 있습니다. 이 데이터를 사용하여 여러 가지 통계적 분석과 배열 조작을 수행하세요.

1. 각 지점에서 6개월 동안 판매된 **사과**, **바나나**, **오렌지**의 월별 평균 판매량을 계산하세요.
2. 각 월마다 전체 지점에서 가장 많이 판매된 과일을 찾아 판매량과 과일명을 출력하세요.
3. 6개월 동안 전체 지점에서 가장 많이 판매된 과일의 총량을 과일별로 구하세요. 어떤 과일이 가장 많이 판매되었는지 확인하세요.
4. 각 지점에서 판매량이 150개 이상인 데이터를 추출하고 그에 해당하는 지점과 과일을 출력하세요.

- **데이터 (**각 지점의 월별 과일 판매량(단위: 개))

    ```python
    import numpy as np

    # 5개 지점의 월별 사과, 바나나, 오렌지 판매량 (5x3x6 배열)
    # 5개의 지점, 3개의 과일(사과, 바나나, 오렌지), 6개월치 데이터
    sales_data = np.array([
        # 지점 1
        [[120, 150, 200, 230, 180, 210],  # 사과 판매량
         [100, 140, 170, 190, 160, 180],  # 바나나 판매량
         [90, 120, 160, 140, 130, 150]],  # 오렌지 판매량

        # 지점 2
        [[80, 90, 110, 140, 130, 120],
         [60, 70, 90, 100, 110, 95],
         [50, 80, 100, 110, 90, 85]],

        # 지점 3
        [[130, 140, 160, 180, 170, 190],
         [110, 120, 130, 150, 140, 160],
         [100, 110, 120, 130, 120, 140]],

        # 지점 4
        [[100, 130, 150, 170, 160, 180],
         [90, 100, 110, 120, 130, 125],
         [80, 110, 140, 130, 120, 110]],

        # 지점 5
        [[90, 110, 130, 140, 120, 130],
         [70, 85, 100, 110, 105, 95],
         [60, 90, 110, 100, 90, 95]]
    ])
    ```

````python
import numpy as np

sales_data = np.array([
    [[120, 150, 200, 230, 180, 210],
     [100, 140, 170, 190, 160, 180],
     [90, 120, 160, 140, 130, 150]],

    [[80, 90, 110, 140, 130, 120],
     [60, 70, 90, 100, 110, 95],
     [50, 80, 100, 110, 90, 85]],

    [[130, 140, 160, 180, 170, 190],
     [110, 120, 130, 150, 140, 160],
     [100, 110, 120, 130, 120, 140]],

    [[100, 130, 150, 170, 160, 180],
     [90, 100, 110, 120, 130, 125],
     [80, 110, 140, 130, 120, 110]],

    [[90, 110, 130, 140, 120, 130],
     [70, 85, 100, 110, 105, 95],
     [60, 90, 110, 100, 90, 95]]
])

fruits = ["사과", "바나나", "오렌지"]
months = ["1월", "2월", "3월", "4월", "5월", "6월"]
````

````python
# -------------------------------------------------
# 1. 각 지점에서 사과, 바나나, 오렌지의 월별 평균 판매량
#    (한 지점 안에서 6개월 동안의 평균, 지점마다 따로)
# -------------------------------------------------
# axis=2: 월 방향으로 평균 -> 결과 shape (5, 3) = (지점, 과일)
avg_per_location_fruit = np.mean(sales_data, axis=2)
print("1. 지점별 과일별 6개월 평균 판매량:")
for loc in range(5):
    print(f"  지점{loc+1}:")
    for f, fruit in enumerate(fruits):
        print(f"    {fruit}: {avg_per_location_fruit[loc, f]:.2f}개")
print()
````

````python
# -------------------------------------------------
# 2. 각 월마다 전체 지점에서 가장 많이 판매된 과일
# -------------------------------------------------
monthly_total_per_fruit = np.sum(sales_data, axis=0)  # (3, 6): 과일x월, 지점 합산
print("2. 월별 가장 많이 판매된 과일:")
for m in range(6):
    fruit_idx = np.argmax(monthly_total_per_fruit[:, m])
    print(f"  {months[m]}: {fruits[fruit_idx]} ({monthly_total_per_fruit[fruit_idx, m]}개)")
print()
````

````python
# -------------------------------------------------
# 3. 6개월 동안 전체 지점에서 과일별 총 판매량
# -------------------------------------------------
total_per_fruit = np.sum(sales_data, axis=(0, 2))  # 지점과 월을 모두 합산
print("3. 과일별 6개월 총 판매량:")
for i, fruit in enumerate(fruits):
    print(f"  {fruit}: {total_per_fruit[i]}개")
best_fruit_idx = np.argmax(total_per_fruit)
print(f"  -> 가장 많이 판매된 과일: {fruits[best_fruit_idx]} ({total_per_fruit[best_fruit_idx]}개)")
print()
````

````python
# -------------------------------------------------
# 4. 판매량이 150개 이상인 데이터의 지점과 과일
# -------------------------------------------------
print("4. 판매량 150개 이상인 데이터:")
locations, fruit_idxs, month_idxs = np.where(sales_data >= 150)
for loc, f, m in zip(locations, fruit_idxs, month_idxs):
    print(f"  지점{loc+1} - {fruits[f]} - {months[m]}: {sales_data[loc, f, m]}개")
````
