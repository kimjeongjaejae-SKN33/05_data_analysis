### [실습 1] 중심 경향 및 산포 측도 계산


아래 배열에서 평균, 중앙값, 최빈값, 분산, 표준편차, 범위를 계산하세요.


```python
	arr = np.array([10, 20, 30, 30, 40, 50, 60])
```

````python
import numpy as np
````

````python
arr = np.array([10, 20, 30, 30, 40, 50, 60])
print("평균: ", np.mean(arr))
print("중앙값: ", np.median(arr))
values, counts = np.unique(arr, return_counts=True)
mode_value = values[np.argmax(counts)]
print("최빈값: ", mode_value)
print("분산: ", np.var(arr, ddof =1))
print("표준편차: ", np.std(arr, ddof =1))
print("범위: ", np.max(arr) - np.min(arr))
````

### [실습 3] 평균 성적 구하기


아래와 같이 학생별 과목 성적 데이터가 주어졌을 때, 각 학생의 전체 평균 성적(수학, 영어, 과학)을 계산한 후, 이 평균값에 대한 분산과 표준편차를 구하세요.


```python
# 학생별 과목 성적 데이터
scores_math = [88, 92, 76, 85, 95, 92, 88, 90, 85, 88]
scores_english = [90, 85, 80, 87, 92, 89, 83, 91, 86, 88]
scores_science = [85, 88, 82, 89, 94, 86, 91, 88, 84, 87]
```

````python
# 학생별 과목 성적 데이터
scores_math = [88, 92, 76, 85, 95, 92, 88, 90, 85, 88]
scores_english = [90, 85, 80, 87, 92, 89, 83, 91, 86, 88]
scores_science = [85, 88, 82, 89, 94, 86, 91, 88, 84, 87]

# 넘파이 배열로 변환
math_arr = np.array(scores_math)
english_arr = np.array(scores_english)
science_arr = np.array(scores_science)

# 학생별 전체 평균 (수학, 영어, 과학)
student_avg = (math_arr + english_arr + science_arr) / 3
print("학생별 평균 성적: ", student_avg)

# 평균값들에 대한 분산과 표준편차
variance = np.var(student_avg, ddof=1)
std_dev = np.std(student_avg, ddof=1)

print("분산: ", variance)
print("표준편차: ", std_dev)
````
