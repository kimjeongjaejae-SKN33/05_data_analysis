    - 2분기 판매량: `[220, 300, 180, 420]` (사과, 배, 오렌지, 바나나)

````python
import numpy as np
````

### [실습 1] 기본 산술 연산


다음 두 배열을 사용하여 각각의 요소별로 덧셈, 뺄셈, 곱셈, 나눗셈을 수행하세요.


```python
arr1 = np.array([10, 20, 30, 40])
arr2 = np.array([1, 2, 3, 4])
```

````python
arr1 = np.array([10, 20, 30, 40])
arr2 = np.array([1, 2, 3, 4])

print("덧셈: ", arr1 + arr2)
print("뺄셈: ", arr1 - arr2)
print("곱셈", arr1 * arr2)
print("나눗셈:", arr1 / arr2)
````

### [실습 2] Dot Product


다음 두 배열 간의 내적 연산을 수행하세요.


```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
```

````python
print("arr1 @ arr2: ", arr1@arr2)
````

### [실습 3] 브로드캐스팅


아래의 2차원 배열과 1차원 배열을 브로드캐스팅하여 더하세요.


```python
arr2d = np.array([[1, 2, 3], [4, 5, 6]])
arr1d = np.array([10, 20, 30])
```

````python
arr2d = np.array([[1, 2, 3], [4, 5, 6]])
arr1d = np.array([10, 20, 30])

print("브로드캐스팅 덧셈:\n " , arr2d + arr1d)
````

### [실습 4] 비교 연산 및 집계 함수 사용


1. 배열에서 20보다 큰 값을 찾아 출력하세요.
2. 배열의 합계, 최대값, 최소값, 평균을 계산하세요.

```python
arr = np.array([10, 20, 30, 40, 50])
```

````python
arr = np.array([10,20,30,40,50])
# 1. 배열에서 20보다 큰 값을 찾아 출력하세요.
high_value = arr[arr>20]
print(high_value)

# 2. 배열의 합계, 최대값, 최소값, 평균을 계산하세요.
print("합계: ", np.sum(arr))
print("최대값: ", np.max(arr))
print("최소값: ", np.min(arr))
print("평균: ", np.mean(arr))
````
