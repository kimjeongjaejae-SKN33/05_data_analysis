```python
# 한글이 포함된 그래프를 깨지지 않게 표시하기 위한 폰트 설정
import platform
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# font_candidates: 운영체제별 한글 폰트 후보
font_candidates = {
    "Darwin": ["AppleGothic", "NanumGothic", "Noto Sans CJK KR", "Noto Sans KR"],
    "Windows": ["Malgun Gothic", "NanumGothic", "Noto Sans CJK KR", "Noto Sans KR"],
    "Linux": ["NanumGothic", "Noto Sans CJK KR", "Noto Sans KR"],
}

# available_fonts: 현재 환경에 설치된 폰트 이름 모음
available_fonts = {font.name for font in fm.fontManager.ttflist}

# platform.system(): 현재 운영체제 이름 확인
system_name = platform.system()
selected_font = None

# 폰트 후보 중 실제 설치된 첫 번째 한글 폰트 선택
for font_name in font_candidates.get(system_name, []):
    if font_name in available_fonts:
        selected_font = font_name
        break

if selected_font:

    # font.family: Matplotlib 기본 글꼴 설정
    plt.rcParams["font.family"] = selected_font
    print("설정된 한글 폰트:", selected_font)
else:
    print("사용 가능한 한글 폰트를 찾지 못했습니다. NanumGothic 또는 Noto Sans KR 설치가 필요할 수 있습니다.")

# axes.unicode_minus=False: 마이너스 기호 깨짐 방지
plt.rcParams["axes.unicode_minus"] = False
```

## Figure와 Axes

https://sigmadream.github.io/practice-matplotlib/note/part_01.html


- Figure(피겨): 전체 그림을 그리는 틀 (캔버스)
- Axes(엑시즈): 그래프가 실제로 그려지는 영역 (데이터와 축을 포함)

**주요명령어**
- 새 figure 생성하기 `plt.figure()`
- 새 axes 생성하기 `plt.axes()`
- 현재 axes 가져오기(get_current_axes) `ax = plt.gca()`
- figure와 그에 속한 axes 동시에 생성하기 `fig, ax = plt.subplots()`

![](https://d.pr/i/nggnqj+)

| 용어         | 의미            | 비유     |
| ---------- | ------------- | ------ |
| Figure     | 전체 그림 영역      | 도화지    |
| Axes       | 실제 그래프 하나의 영역 | 그래프 칸  |
| Axis       | x축, y축        | 눈금자    |
| Tick       | 축의 눈금         | 눈금 선   |
| Tick label | 눈금에 붙은 숫자/글자  | 눈금 값   |
| Spine      | 그래프 테두리       | 액자 테두리 |
| Legend     | 범례            | 설명표    |


```python
# figure == 전체 틀
# axes == 실제 그래프가 그려지는 영역

# 새 Figure를 만들고 선/막대 그래프 그리기

# figsize=(8,6) : 피겨 가로 8인치, 세로 6인치
# facecolor='pink' : 배경색 핑크
plt.figure(figsize=(8,6), facecolor='pink')

# 현재 활성화된 axes 가져오기, 없으면 생성
ax = plt.gca() #get current axes 약자
X = [1,2,3]
Y = [2,4,6]

ax.plot(X,Y, marker = 'o') # 선그래프
ax.bar(X,Y, color='orange', alpha=0.5, width=0.5)
plt.title("Figure - Axes")
plt.show()
```

```python
# plt.subplots()
# - figure 1개와 여러 axes를 함께 생성
# - figure 내부에 여렁 axes가 배치되고 [row. col] 형식으로 접근, 제어

# figure 내에 axes 2행 3열 생성
# figure의 크기는 가로 8인치, 세로 6인치
fig, ax = plt.subplots(nrows=2, ncols=3, figsize = (8,6))

print(ax.shape)

X = [1,2,3]
y = [2,4,6]
ax[0,0].plot(X,y)
ax[0,1].plot(X,y)
ax[0,2].plot(X,y)

# Axes.bar() : 막대 그래프 생성
ax[1,0].plot(X,y)

# Axes.scatter(): 산점도 그래프 생성
ax[1,1].scatter(X,y)

# Axes.axis("off")
ax[1,2].axis("off")

plt.show()
```

## plot

```python
# 선 그래프 생성 + 선 꾸미기
plt.plot(X,y,
         color='red',
         linestyle='dashed',
         linewidth=3,
         marker='o',
         markersize=20,
         label = 'my line plot'
)
plt.legend(loc = 'lower center')

# 축 이름 지정
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# plt.xticks(): x축 눈금 위치/라벨 설정
plt.xticks([1,2,3],  ['A', 'B', 'C'])

# plt.yticks(): y축 눈금 위치/라벨 설정
# rotation =45 : 글자 45도 회전
plt.yticks(ticks=range(0,11), rotation = 45)

# pltt.xlim(), plt.ylim() : x/y 축 표시 범위 제한
plt.xlim(2,3)
plt.ylim(2,8)

plt.show()
```

```python
import numpy as np
X = np.arange(1,10)
Y = X * 2
Z = X * 4

plt.figure(figsize=(12,8))
plt.plot(X,Y, color ='red', label = 'X, Y')
plt.plot(X,Z, color ='purple', label = 'X, Z')

plt.title("선 그래프 2개")
plt.legend(loc = 'upper left')
plt.show()
```

```python
fig, ax = plt.subplots(nrows= 1, ncols = 2, figsize = (12,6))
ax[0].plot(X,Y, color ='red', linewidth =2 ,label = 'X, Y')
ax[1].bar(X, Z, color='orange', width =0.5 ,label = 'X, Z')

ax[0].legend(loc = 'upper left')
ax[1].legend(loc = 'upper left')

plt.show()
```
